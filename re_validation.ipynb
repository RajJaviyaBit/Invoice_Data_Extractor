````python
def validate_response(final: dict):
    irn_val = "^[A-Za-z0-9]{64}$"
    ack_no_val = "^[0-9]{15}$"
    # date_val = "^[0-9]{2}/[0-9]{2}/[0-9]{2}$"
    amount_val = "[-+]?([0-9]*[.])?[0-9]+([eE][-+]?[0-9])?"
    dealer_code_val= "^[A-Z]{1}[0-9]{4}$"
    hiib_misp_code_val = "^HIIB(-| )?MHY(-| )?[0-9]{4}$"
    pincode_val = "^[0-9]{6}$"
    state_code_val = "^([0-9]{2})|([0-9]{1})$"
    MSME_val = "^UDYAM-[A-Z]{2}-[0-9]{2}-[0-9]{7}$"
    ifsc_val = "^[A-Z]{4}0[A-Z0-9]{6}$"
    PAN_val = "^[A-Z]{5}[0-9]{4}[A-Z]{1}$"
    gst_val = "^[0-9]{2}[A-Z]{5}[0-9]{4}[A-Z]{1}[1-9A-Z]{1}Z[0-9A-Z]{1}$"
    sac_val = "^[0-9]{6}$"
    quant_val = "[0-9]+"
    
    final['irn'] = final['irn'].replace("-", "")

    if re.fullmatch(pattern=irn_val, string=str(final["irn"])): pass       
    else: final["irn"] = "null"

    if re.fullmatch(pattern=ack_no_val,string=str(final["ack_no"])): pass
    else: final["ack_no"] = "null"
   

    if re.fullmatch(pattern=amount_val,string=str(final['taxable_value'])): pass
    else: final['taxable_value'] = "null"


    if re.fullmatch(pattern=amount_val,string=str(final['cgst_amount'])): pass
    else: final['cgst_amount'] = "null"


    if re.fullmatch(pattern=amount_val,string=str(final['sgst_utgst_amount'])): pass
    else: final['sgst_utgst_amount'] = "null"

    if re.fullmatch(pattern=amount_val,string=str(final['igst_amount'])): pass
    else: final['igst_amount'] = "null"


    if re.fullmatch(pattern=amount_val,string=str(final['total_invoice_value'])): pass
    else: final['total_invoice_value'] = "null"
  

    if re.fullmatch(pattern=dealer_code_val,string=str(final['dealer_code'])): pass
    else: final['dealer_code'] = "null" 

    if re.fullmatch(pattern=hiib_misp_code_val, string=str(final['hiib_misp_code'])): pass
    else: final['hiib_misp_code'] = "null"

    if re.fullmatch(pattern="^[0-9]+$", string=str(final['account_no'])): pass
    else: final['account_no'] = "null"

    if re.fullmatch(pattern=ifsc_val, string=str(final['bank_ifsc'])): pass
    else: final['bank_ifsc'] = "null"

    if re.fullmatch(pattern="^[0-9]{9}$", string=str(final['micr_code'])): pass
    else: final['micr_code'] = "null"

    if re.fullmatch(pattern=gst_val, string=str(final['hiib_gstin'])): pass
    else: final['hiib_gstin'] = "null"

    if re.fullmatch(pattern=gst_val, string=str(final['dealer_gstin'])): pass
    else : final['dealer_gstin'] = "null"

    if re.fullmatch(pattern=pincode_val, string=str(final['hiib_pincode'])): pass
    else: final['hiib_pincode'] = "null"

    if re.fullmatch(pattern=state_code_val, string=str(final['dealer_state_code'])): pass
    else: final['dealer_state_code'] = "null"


    if re.fullmatch(pattern=MSME_val, string=str(final['msme_code'])): pass
    else: final['msme_code'] = "null"


    if re.fullmatch(pattern=PAN_val, string=str(final['dealer_pan'])): pass
    else: final['dealer_pan'] = "null"


    if re.fullmatch(pattern=pincode_val, string=str(final['dealer_pincode'])): pass
    else: final['dealer_pincode'] = "null"


    if re.fullmatch(pattern=state_code_val, string=str(final['hiib_state_code'])): pass
    else: final['hiib_state_code'] = "null"


    if re.fullmatch(pattern=sac_val, string=str(final['sac'])): pass
    else: final['sac'] = "null"


    if re.fullmatch(pattern=pincode_val, string=str(final['consigner_pincode'])): pass
    else: final['consigner_pincode'] = "null"


    if re.fullmatch(pattern=pincode_val, string=str(final['buyer_pincode'])): pass
    else: final['buyer_pincode'] = "null"


    if re.fullmatch(pattern=quant_val,string=str(final['quantity'])): pass
    else: final['quantity'] = "null"
    
    return final
````

````python
def response_process(answer):
    # remove unnecessary part from response
    answer = re.sub("\n", "", answer)
    answer = re.sub("json", "", answer)
    answer = re.sub("```", "", answer)
    # jsonify the response
    final = json.loads(answer)
    return final
````
